### Ejemplo Viscosidad (Problema 2)

_G. Raush_

2022

Un fina placa separa dos fluidos de viscosidades $\mu_1$ y $\mu_2$ . El espacio de separacion entre placas fijas y la movil es diferente $h_1$ y $h_2$ . El área de contacto $A$ es la misma para ambos fluidos. Toma la hipñótesis de un perfil de velocidad lineal, $\dot{\gamma}$ constante. 

a) Calcular la fuerza necesaria para conseguir una velocidad $V$

b) Calcular la relación de viscosidades corresondiente a la geometria dad acon tal de conseguir un régimen estable en la posición de la placa movil.
